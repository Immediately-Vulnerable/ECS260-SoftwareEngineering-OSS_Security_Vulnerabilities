```python
import csv 

# Read node list in the dependency graph
npm_dir = "/content/drive/Shareddrives/ECS260-group5/dataset/npm/"

with open(npm_dir + 'node_list.txt', "r") as f:
  node_list = f.read().splitlines()

def getstuff(filename, criterion):
    with open(filename, "r") as csvfile:
      datareader = csv.reader(csvfile)

      header = next(datareader)
      header.append('isolated')  # Add isolation flag header
      yield header  # yield the header row

      for row in datareader:
        platform = row[1]
        if platform == criterion: # does platform match the given criterion (='NPM')?
          row.pop(55) #pop garbage column which causes mismatch

          # is this project isolated?
          proj_id = row[0]
          if proj_id in node_list: isIsolated = False
          else: isIsolated = True

          # Set isolation flag
          row.append(str(isIsolated))

          # yield row
          yield row                

with open(dir +'(NPM_extracted)' + fname, 'w') as f:
  # create the csv writer
  writer = csv.writer(f)

  # write a row to the csv file
  for row in getstuff(dir + fname, 'NPM'):
    writer.writerow(row)
```

```python
org_file = dir + fname
ext_file = dir +'(NPM_extracted)' + fname

with open(org_file) as f:
    dat_num = sum(1 for line in f)
    print('Org file data rows: ', dat_num)

with open(ext_file) as f:
    dat_num = sum(1 for line in f)
    print('Extracted file data rows: ', dat_num)  
```
